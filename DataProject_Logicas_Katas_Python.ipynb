1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python

def contar_frecuencia():
    cadena = input("Introduce una cadena de texto").lower().strip()

    """Funcion que Recibe una cadena y devuelve un diccionario con la frecuencia de cada letra sin los espacios.
    
    Args:
        cadena (str): Cadena de texto

    Returns:
        _dict_: Devuelve un diccionario con la frecuencia de cada letra sin los espacios.
    """
    
    frecuencias = {}
    
    for indice,letra in enumerate(cadena):
        if letra == " ":
            continue

        if letra in frecuencias:
            frecuencias[letra] += 1
        else:
            frecuencias[letra] = 1
            
    return print(f"Frecuencia de letras: {frecuencias}")

# Invocamos la funcion
contar_frecuencia()
```

devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def filtrar_palabras_por_objetivo():
    """
    Toma una lista de palabras y una palabra objetivo por input,
    y devuelve las palabras que contienen la palabra objetivo.

    Args:
        enentrada_lista (str): Introduce palabras separadas por espacios por un input
        lista_palabras (list,str): Convierte la cadena en una lista
        palabra_objetivo (str): Introduce la palabra objetivo a buscar por un input
    
    Returns:
        list : Retorna una lista con las palabra que incluyen la palabra_objetivo
    """
    entrada_lista = input("Introduce palabras separadas por espacios: ")
    lista_palabras = entrada_lista.split()  # Convierte la cadena en una lista
    
    palabra_objetivo = input("Introduce la palabra objetivo a buscar: ")
    
    palabras_encontradas = []
    for palabra in lista_palabras:
        if palabra_objetivo in palabra:
            palabras_encontradas.append(palabra)
            
    return f"Las palabras_encontradas con la palabra objetivo {palabra_objetivo} son : {palabras_encontradas}"

filtrar_palabras_por_objetivo()

# --- Ejemplo de uso ---
# Supongamos que se introduce las palabras: "sol soleado girasol luna tierra insolación"
# objetivo: "sol"

```

4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función  map()

```python
def calcular_diferencia_listas(lista1, lista2):
    """
    Calcula la diferencia (lista1[i] - lista2[i]) de dos listas
    utilizando map() y lambda.

    Arg:
        lista1,lista2 (list): listas de valores a elegir

    return:
        list: La diferencia entre las dos listas elegidas
    """
    
    diferencia = list(map(lambda p, q: p - q, lista1, lista2))
    return diferencia


lista1 = [10, 20, 30, 40]
lista2 = [2, 5, 10, 15]

print(f"Lista 1: {lista1}")
print(f"Lista 2: {lista2}")

resultado= calcular_diferencia_listas(lista1,lista2)


print(f"La diferencia entre las dos listas es: {resultado}")
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por 
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual 
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver 
una tupla que contenga la media y el estado.

```python
def evaluar_rendimiento(notas, nota_aprobado=5):
    """
        Calcula la media de una lista de notas y determina si está aprobado o suspenso.

    Args:
        notas (list): Lista de números (int o float).
        nota_aprobado (int o float, opcional): Nota mínima para aprobar. Por defecto es 5.

    Returns:
        tuple: (media, estado) donde media es float y estado es "aprobado" o "suspenso".
    """
   
    if not notas:
        return 0.0, "suspenso"
    
    media = sum(notas) / len(notas)
    
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
        
    return (media, estado)


notas_1 = [4, 8, 6]
evaluar_rendimiento(notas_1)
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial_recursivo(n):
    """
    Calcula el factorial de un número utilizando recursión.

    Summary:
        La función se llama a sí misma reduciendo el problema en n-1 
        hasta llegar al caso base (0 o 1), donde la recursión se detiene.

    Args:
        n (int o float ): Un número entero no negativo.

    Returns:
        int o float: Retorna el factorial del numero indicado. 
        Si se indica un negativo devolvera el valor de 1
    """
    if n <= 1:
        return 1
    
    return n * factorial_recursivo(n - 1)

factorial_recursivo(-1)
```

10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una 
excepción personalizada y maneja el error adecuadamente.

```python
def calcular_promedio(lista):
    """
    Calcula el promedio de una lista de números de manera segura.
    En Caso que lista esta vacia o se indica un valor no numerico lanza expeción y error.
    
    Args:
        lista (list): Una colección de números (int o float).

    Returns:
        float: El promedio de los números si la operación es exitosa.
        str: Un mensaje de error si la lista está vacía o contiene datos no numéricos.
    """
    try:
        if not lista:
            return "Error: La lista está vacía, no se puede calcular promedio."
        promedio = sum(lista) / len(lista)
        return promedio
    
    except TypeError:
        return "Error: La lista contiene valores no numéricos."

# Lista con datos
numeros = [10, 20, 30, 40, 50]
resultado = calcular_promedio(numeros)
print(f"Lista con datos -> Promedio: {resultado}")

# Lista vacía
lista_vacia = []
resultado_vacio = calcular_promedio(lista_vacia)
# Se maneja el error sin que el programa se detenga
print(f"Lista vacia -> {resultado_vacio}")

#Lista con datos no numéricos
lista_invalida = [10, "a", 30]
resultado_invalido = calcular_promedio(lista_invalida)
print(f"Lista no valida -> {resultado_invalido}")
```

13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en 
mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función  map()

```python
def letras_unicas(cadena):
    """
    Funcion que convierte un conjunto decaracteres en una lista de tuplas 
    con cada letra en (mayúscula, minúscula) sin repetir letras.
        
    Args:
        cadena (str): Conjunto de caracteres de un texto dado.
        
    Returns:
        list: Lista de tuplas con cada letra en Mayusculas y Minusculas sin contar espacios.
    """
    
    letras = []
    for c in cadena.replace(" ", ""):
        if c.lower() not in letras:
            letras.append(c.lower())
        
    lista_tupla = map(lambda l: (l.upper(), l.lower()), letras)
        
    return list(lista_tupla)


texto = "Hola hola Adios adios"
print(letras_unicas(texto))

```

29.  Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres  con el 
carácter '#', excepto los últimos cuatro.

```python
def enmascarar_cadena(valor):
    """
    Función que convierte una variable en una cadena de texto y enmascara todos los caracteres con el carácter '#', excepto los últimos cuatro.
    
    Args:
        valor (str): variable que se convierte en cadena de texto

    Returns:
        _str_: cadena de texto con # menos los ultimos 4
    """

    texto = str(valor)
        
    if len(texto) <= 4:
        return texto
    
    enmascarado = '#' * (len(texto) - 4) + texto[-4:]
    
    return enmascarado


print(enmascarar_cadena("Convierte una variable"))
print(enmascarar_cadena(987654321))         
print(enmascarar_cadena("tres"))            
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras 
pero en diferente orden.

```python
def es_anagrama(palabra1, palabra2):
    """Función que determina si dos palabras son anagramas

    Args:
        palabra1 (str): Primera palabra para comparar si es anagrama de la palabra2
        palabra2 (str): Segunda palabra para comparar si es anagrama de la palabra1

    Returns:
        bool: Retorna True o False dependiendo si es anagrama o no. 
    """

    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()
    
    if len(p1) != len(p2):
        return False
    
    return sorted(p1) == sorted(p2)


print(es_anagrama("Roma", "Amor")) 
print(es_anagrama("Hola", "Ola"))  
print(es_anagrama("fresa", "frase")) 
```

31. Crea una función que solicite al usuario ingresar una lista de nombres y luego solicite un nombre para buscar en 
esa lista. Si el nombre está en la lista, se imprime un mensaje indicando que fue encontrado, de lo contrario, se 
lanza una excepción.

```python
def gestionar_busqueda_nombres():
    """
    Funcion que solicita ingresar una lista de nombres y un nombre a buscar.

    Args:
        (str): Los nombre a ingresar se transtormar a capitaliza() para que todos sean iguales

    Returns:
        (str): Retorna un texto si fue encontrado el nombre y otro si no. 
    """
    
    entrada = input("Ingrese nombres separados por coma: ")
    print(f"La lista original de nombres es {entrada}")

    lista_nombres = []
    for nombre in entrada.split(','):
        lista_nombres.append(nombre.strip().capitalize())
    
    print(f"La lista de nombres es: {lista_nombres}")
    
    nombre_buscado = input("Ingrese el nombre a buscar: ").capitalize()
    print(f"El nombre buscado es {nombre_buscado}")

    if nombre_buscado in lista_nombres:
        print(f"El nombre {nombre_buscado} fue encontrado en la lista.")
    else:
       
        print(f"El nombre {nombre_buscado} no existe en la lista.")

gestionar_busqueda_nombres()
```

```python
def buscar_puesto_empleado(nombre_completo, lista_empleados):
    """
    Funcion que busca el puesto de un empleado por su nombre completo en una lista de diccionarios
    y devuelve el valor asociado a la clave 'puesto' o mensaje "no trabaja aqui".

    Args:
        nombre_completo (str): El nombre completo del empleado a buscar.
        lista_empleados (list): Una lista de diccionarios que contiene 'nombre' (str) y 'puesto'(str).

    Returns:
        str: Retorna el puesto del empleado elegido si se encuentra, 
        y si no lo encuentra,un mensaje indicando que no trabaja en la empresa.
    """
    for empleado in lista_empleados:
        if empleado['nombre'] == nombre_completo:
            return f"{nombre_completo} tiene el puesto de {empleado['puesto']}"
        
    return f"{nombre_completo} no trabaja aquí."


empleados = [
        {'nombre': 'Ana Perez', 'puesto': 'Gerente'},
        {'nombre': 'Juan Diaz', 'puesto': 'Programador'},
        {'nombre': 'Maria Gomez', 'puesto': 'Administradora'}
    ]


busqueda1 = buscar_puesto_empleado('Juan Diaz', empleados)
print(busqueda1)
print('-' * 30)
busqueda2 = buscar_puesto_empleado('Pedro Ruiz', empleados)
print(busqueda2)
```

34. Crea la clase  Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: 
crecer_tronco ,  nueva_rama ,  crecer_ramas ,  quitar_rama  e  info_arbol . El objetivo es implementar estos métodos para 
manipular la estructura del árbol.

Código a seguir:
 1- Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
 2- Implementar el método  crecer_tronco  para aumentar la longitud del tronco en una unidad.
 3- Implementar el método  nueva_rama  para agregar una nueva rama de longitud 1 a la lista de ramas.
 4- Implementar el método  crecer_ramas  para aumentar en una unidad la longitud de todas las ramas existentes.
 5- Implementar el método  quitar_rama  para eliminar una rama en una posición específica.
 6- Implementar el método info_arbol  para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.

Caso de uso:
 1- Crear un árbol.
 2- Hacer crecer el tronco del árbol una unidad.
 3- Añadir una nueva rama al árbol.
 4- Hacer crecer todas las ramas del árbol una unidad.
 5- Añadir dos nuevas ramas al árbol.
 6-Retirar la rama situada en la posición 2.
 7- Obtener información sobre el árbol.

```python
class Arbol:
    """
    Clase que representa la evolucion de un arbol
    """
    def __init__(self):
        """
        Constructor de la clase Arbol.
        
        Args:
            tronco (int): valor incicial de longitud del arbol 1
            ramas(list): Lista vacia para ir modificando las ramas
                """
        self.tronco = 1
        self.ramas = []

    def crecer_tronco(self):
        """
        Metodo para aumentar la longitud del tronco en una unidad.
        
        Args:
            tronco (int): Incremente una unidad (int) el tronco de la clase arbol
        """
        self.tronco += 1
        print(f"El tronco ha crecido. Longitud actual: {self.tronco}")

    def nueva_rama(self):
        """
        Método para agregar una nueva rama de longitud 1 a la lista de ramas.

        Args:
            ramas (int): 1 rama nueva que añade la lista de ramas
        """
        self.ramas.append(1)
        print(f"Nueva rama añadida. Total ramas: {len(self.ramas)}")

    def crecer_ramas(self):
        """
        Método para aumentar en una unidad la longitud de todas las ramas existentes.
        
        Args:
            ramas (int): 1 rama nueva a las ya existentes 
        """
        self.ramas = [rama + 1 for rama in self.ramas]
        print("Todas las ramas existentes han crecido:", self.ramas)

    def quitar_rama(self, posicion):
        """
        Método  quitar_rama  para eliminar una rama en una posición específica.

        Args:
            posicion (int): Indica la posicion elegida para ser eliminada
        """
        if 0 <= posicion < len(self.ramas):
            eliminada = self.ramas.pop(posicion)
            print(f"Rama en posición {posicion}  eliminada.")
        else:
            print("Posición de rama inválida.")

    def info_arbol(self):
        """
        Método para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.
        """       
        print("------------Informacion del Arbol-----")
        print(f"Longitud tronco: {self.tronco}")
        print(f"Número de ramas: {len(self.ramas)}")
        print(f"Longitudes de ramas: {self.ramas}")
        

# --- Caso de uso ---
# 1. Crear un árbol.
mi_arbol = Arbol()

# 2. Hacer crecer el tronco del árbol una unidad.
mi_arbol.crecer_tronco()

# 3. Añadir una nueva rama al árbol.
mi_arbol.nueva_rama()

# 4. Hacer crecer todas las ramas del árbol una unidad.
mi_arbol.crecer_ramas()

# 5. Añadir dos nuevas ramas al árbol.
mi_arbol.nueva_rama()
mi_arbol.nueva_rama()

# 6. Retirar la rama situada en la posición 2.
mi_arbol.quitar_rama(2)

# 7. Obtener información sobre el árbol.
mi_arbol.info_arbol()
```

36. Crea la clase  UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta 
corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y 
agregar dinero al saldo.
Código a seguir:
1- Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante  True  y  False .
2- Implementar el método  retirar_dinero  para retirar dinero del saldo del usuario. Lanzará un error en caso de no poder hacerse.
3-Implementar el método  transferir_dinero  para realizar una transferencia desde otro usuario al usuario actual. Lanzará un error en caso de no poder hacerse.
4-Implementar el método  agregar_dinero  para agregar dinero al saldo del usuario.

Caso de uso:
1- Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
2- Agregar 20 unidades de saldo de "Bob".
3- Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
4- Retirar 50 unidades de saldo a "Alicia".


```python
class UsuarioBanco:
    """
    Clase que epresenta a un usuario de un banco con nombre, saldo y tipo de cuenta.
    """

    def __init__(self, nombre, saldo, tiene_cuenta_corriente):
        """
        Método constructor de la Clase UsuariBanco.

        Args:
            nombre (str) : nombre del usuario
            saldo (floar): saldo del usuario
            tiene_cuenta_corriente (bool): Indica True si tiene y False si no tiene cuenta corriente
        """
        self.nombre = nombre
        self.saldo = saldo
        self.tiene_cuenta_corriente = tiene_cuenta_corriente

    def agregar_dinero(self, cantidad):
        """
        Función que suma una cantidad al saldo actual del usuario.

        Arg:
            cantidad (float): Valor que suma al saldo del usuario.
        
        Returns:
            cantidad (float): Retorna la cantidad sumada, el nombre y el total del saldo del usuario.
        """
        self.saldo += cantidad
        print(f"Se han agregado {cantidad} unidades. Nuevo saldo de {self.nombre}: {self.saldo}")

    def retirar_dinero(self, cantidad):
        """
        Resta una cantidad al saldo si hay fondos suficientes; si no, muestra un error.
        
        Arg:
            cantidad (float): Valor que resta el saldo del usuario.
        
        Returns:
            cantidad (float): Retorna la cantidad restada, el nombre y el total del saldo del usuario.
            condicionales : Condiciona si se puede o no realizar el retiro de dinero, según el saldo disponible.
        """
        if cantidad <= self.saldo:
            self.saldo -= cantidad
            print(f"Retiro exitoso. Nuevo saldo de {self.nombre}: {self.saldo}")
        else:
            print(f"ERROR: {self.nombre} no tiene saldo suficiente para retirar {cantidad}")

    def transferir_dinero(self, otro_usuario: 'UsuarioBanco', cantidad: float):
        """
        Funcion que Transfiere o Mueve saldo desde otro_usuario hacia este usuario tras validar saldo.

        Arg:
            otro_usuario (str): Usuario a quien se realiza la transferencia
            cantidad (float): Valor que se le transfiere a otro usuario.
        
        Returns:
            cantidad (float): Retorna la cantidad Transferida, el usuario que la realiza y del usuario que la recibe.
            condicionales : Condiciona si se puede o no realizar la transferncia de dinero, según el saldo disponible.
        """
        if cantidad <= otro_usuario.saldo:
            otro_usuario.saldo -= cantidad
            self.saldo += cantidad
            print(f"Transferencia exitosa: {otro_usuario.nombre} envió {cantidad} a {self.nombre}")
        else:
            print(f"ERROR de transferencia: {otro_usuario.nombre} no tiene saldo suficiente.")

# --- CASO DE USO ---

# 1. Crear dos usuarios
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

# 2. Agregar 20 unidades de saldo a Bob
bob.agregar_dinero(20)

# 3. Hacer una transferencia de 80 unidades desde Bob a Alicia
# (Bob tiene 70, por lo que este método imprimirá el error directamente)
alicia.transferir_dinero(bob, 80)

# 4. Retirar 50 unidades de saldo a Alicia
alicia.retirar_dinero(50)

print("\n#### Comprobacion ####")
print(alicia.__dict__)
print(bob.__dict__)
```

37. Crea una función llamada  procesar_texto  que procesa un texto según la opción especificada:  contar_palabras , 
reemplazar_palabras ,  eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro 
de la función  procesar_texto .

Código a seguir: 
- 1.  Crear una función  contar_palabras  para contar el número de veces que aparece cada palabra en el texto. Tiene 
que devolver un diccionario. 
- 2. Crear una función  reemplazar_palabras  para remplazar una  palabra_original  del texto por una  palabra_nueva . Tiene 
que devolver el texto con el remplazo de palabras. 
- 3.  Crear una función  eliminar_palabra  para eliminar una palabra del texto. Tiene que devolver el texto con la palabra 
eliminada. 
- 4. Crear la función  procesar_texto  que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un 
número de argumentos variable según la opción indicada.

Caso de uso:
Comprueba el funcionamiento completo de la función  procesar_texto

```python
def contar_palabras(texto):
    """Función que cuenta el número de veces que aparece cada palabra en el texto.

    Args:
        texto (str): texto que introducimos para recuento

    Returns:
        (str,int): Retorna la palabra del texto y el numero de veces que aparece
    """
    palabras = texto.lower().split()
    conteo = {}
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1
    return conteo


def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """Función que reemplaza una palabra por otra utilizando el método replace.

    Args:
        texto (str): texto que introcucimos
        palabra_original (str): palabra que indicamos para reemplazar
        palabra_nueva (str): palabra que sustituye a palabra_original

    Returns:
        (str): Nuevo texto con la palabra reemplazada
    """
    return texto.replace(palabra_original, palabra_nueva)

def eliminar_palabra(texto, palabra_a_eliminar):
    """Función que elimina una palabra del texto y devuelve el texto con la palabra eliminada

    Args:
        texto (str): texto que introcucimos
        palabra_a_eliminar (str): palabra que indicamos para sustituir

    Returns:
        (str): Devuelve el texto con la palabra eliminada
    """
    return texto.replace(palabra_a_eliminar, "")


def procesar_texto(texto, opcion, *args):
    """
    Función que toma un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.
    """
    if opcion == "contar":
        return contar_palabras(texto)
    
    elif opcion == "reemplazar":
        if len(args) == 2:
            return reemplazar_palabras(texto, args[0], args[1])
        else:
            return "Error: Faltan argumentos para reemplazar."
            
    elif opcion == "eliminar":
        if len(args) == 1:
            return eliminar_palabra(texto, args[0])
        else:
            return "Error: Falta la palabra a eliminar."
            
    else:
        return "Opción no reconocida."

# --- CASO DE USO ---

texto_ejemplo = "Comprueba el funcionamiento completo de la función procesar_texto"
print("El texto ejemplo es:", texto_ejemplo)

# 1. Contar
print("Resultado contar_palabras:", procesar_texto(texto_ejemplo, "contar"))

# 2. Reemplazar
print("Resultado reemplazar_palabras:", procesar_texto(texto_ejemplo, "reemplazar", "Comprueba", "Comprobado"))

# 3. Eliminar
print("Resultado eliminar_palabra:", procesar_texto(texto_ejemplo, "eliminar", "Comprueba el"))
```

40. Escribe una función que tome dos parámetros:  figura  (una cadena que puede ser  "rectangulo" ,  "circulo"  o 
"triangulo" ) y  datos  (una tupla con los datos necesarios para calcular el área de la figura).

```python
def calcular_area(figura, datos):
    """
    Función que calcula el área de una figura geométrica.
    
    Args:
        figura (str): "rectangulo", "circulo" o "triangulo".           
        datos (tuple): Datos necesarios para calcular el área de la figura. 
            - Fórmula del rectangulo: base * altura  
            - Fórmula del circulo: pi * r^2       
            - Fórmula del Triangulo: (base * altura) / 2                     
    Returns:
        float: El área de la figura seleccionada.
    """
    figura = figura.lower()
    
    if figura == "rectangulo":
        base, altura = datos
        return float(base * altura)
        
    elif figura == "circulo":
        radio = datos[0]
        PI = 3.14159
        return round(PI * (radio ** 2),2)
        
    elif figura == "triangulo":
        base, altura = datos
        return (base * altura) / 2
        
    else:
        return "Figura no reconocida"


area_rect = calcular_area("rectangulo", (10, 5))
print(f"Área rectángulo: {area_rect}")

area_circ = calcular_area("circulo", (3,))
print(f"Área círculo: {area_circ}")

area_tri = calcular_area("triangulo", (4, 6))
print(f"Área triángulo: {area_tri}")
```
